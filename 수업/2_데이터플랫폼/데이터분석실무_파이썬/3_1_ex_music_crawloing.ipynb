## 1. 멜론 차트100 제목, 가수, 랭킹 출력

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np

ser = Service('../chromedriver/chromedriver.exe')
driver = webdriver.Chrome(service=ser)
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
songs = soup.select('tbody > tr')
songs
```

```python
song_data = [[]]

for song in songs:
    _ = song.select('td > div > div > div > span > a')
    title = _[0].text
    muse = _[1].text
    rank = song.select('td > div > span.rank')[0].text
    
    song_data.append(['melon', rank, title, muse])
```

```python
_ = pd.DataFrame(song_data, columns=['서비스','순위','타이틀','가수'])
_.to_excel('./filesmelon_class.xlsx', index=False, sheet_name='Jan 3rd chart 100')
```

### #실습 - 벅스 차트100 제목, 가수, 랭킹 출력

```python
driver = webdriver.Chrome(service=ser)
url = 'https://music.bugs.co.kr/chart'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('tbody > tr')[:100]
songs_data = []
for song in songs:
    rank = song.select('td > div.ranking > strong')[0].text
    title = song.select('td.check > input')[0]['title']
    muse = song.select('td > a.btnActions')[0]['artist_disp_nm']
    songs_data.append(['burgs', rank, title, muse])
```

```python
_ = pd.DataFrame(songs_data, columns=['서비스','순위','타이틀','가수'])
_.to_excel('./filesbugs_class.xlsx', index=False, sheet_name='Jan 3rd chart 100')
```

### #실습 - 지니 차트 50 제목, 가수, 랭킹 출력

```python
url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('tbody > tr')
songs_data = []

for song in songs:
    title = song.select('td >a.title')[0].text.strip()
    rank = song.select('td.number')[0].text.split('\n')[0]
    muse = song.select('td > a.artist.ellipsis')[0].text
    
    songs_data.append(['genie', rank, title, muse])
```

```python
_ = pd.DataFrame(songs_data, columns=['서비스','순위','타이틀','가수'])
_.to_excel('./filesgenie_class.xlsx', index=False, sheet_name='Jan 3rd chart 50')
```
