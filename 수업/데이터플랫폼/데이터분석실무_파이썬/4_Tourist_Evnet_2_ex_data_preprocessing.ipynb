# 3. 전처리 함수화

```python
def create_kto_data(yy, mm):
    # 엑셀 경로 지정
    file_path = './files/kto_{}{}.xlsx'.format(yy, mm)
    
    # 파일 불러오기
    df = pd.read_excel(file_path, header=1, skipfooter=4, usecols='A:G')
    
    # "기준년월" 컬럼 할당
    df['기준년월']='{}-{}'.format(yy, mm)
    
    # 국적 컬럼에서 대륙 제거하고 국가만 남기기.
    ignore_list = ['아시아주','미주','구주','대양주','아프리카주','기타대륙','교포소계']
    condition = df['국적'].isin(ignore_list)==False
    df_country = df[condition].reset_index(drop=True)
    
    # "대륙" 컬럼 추가
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3\
            + ['아프리카']*2 + ['기타대륙'] + ['교포']
    df_country['대륙'] = continents
    
    # 국가별 "관광객비율(%)" 컬럼 추가
    df_country['관광객비율(%)'] = round(df_country['관광'] / df_country['계'] * 100)
    
    # "전체비율(%)" 컬럼 추가
    tourist_sum = df_country['관광'].sum()
    df_country['전체비율(%)'] = round(df_country['관광'] / tourist_sum * 100)
    
    return df_country
    
```

```python
_ = create_kto_data(2019, '11')
_.head()
```

# 4. 함수를 이용한 파일 병합

```python
df = pd.DataFrame()

for yy in range(2010,2021):
    for mm in range(1,13):
        yy, mm = str(yy), str(mm).zfill(2)
        
        try:
            temp = create_kto_data(yy, mm)
            df = df.append(temp, ignore_index=True)
        except:
            pass
        
```
