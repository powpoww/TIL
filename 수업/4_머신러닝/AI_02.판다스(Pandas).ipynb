## 1.2 Dataframe의 Data 선택, 필터링

- Dataframe[ ] 의 연산자는 Numpy[ ] 와는 다르다.
- Dataframe[ ] 에 입력 가능한 값

    - 컬럼명(또는 컬럼명 리스트)
    - 불린 인덱스
    - 행의 슬라이싱 연산
- loc[ , ] : 명칭 기반 인덱싱 연산자

    - 행 : Dataframe의 index (또는 index 리스트)
    - 열 : Dataframe의 컬럼명 (또는 컬럼명 리스트)
- iloc[ , ] : 위치 기반 인덱싱 연산자

    - 행 : Dataframe의 행위치
    - 열 : Dataframe의 열위치


```python
# df.loc
df.loc[((df.Sex=='male') & (df.Age>50)), 'Pclass'].head()
```

```python
# df.iloc
df.iloc[5:10, 3:6]
```

```python
# sort
df.sort_values(by=['Age','Fare'], ascending=False).head()
```

## 1.3 Groupby

```python
df.loc[:,['Fare', 'Age']].mean()
```

```python
df.groupby(by='Pclass').mean()
```

```python
# aggregation
df.groupby(by='Sex')[['Age','Fare']].agg(['min', 'max', 'std'])
```

## 1.4 NaN 처리

```python
# isna
df.isna().sum()
```

```python
# without NaN
df.Age[~df.Age.isna()].head()
```

```python
# fillna
temp = df.Embarked.fillna('N')
print(temp[df.Embarked.isna()])

# method - ffill
temp = df.Embarked.fillna(method='ffill')
print(temp[df.Embarked.isna()])
```

```python
# dropna
df.dropna()
```
